```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
X, y = make_blobs(
    n_samples=200,
    n_features=2,
    centers=5,
    cluster_std=2,
    random_state=42,
)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X[:, 0], X[:, 1], c=y, marker='o', s=100, edgecolors='brown')
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

knn4 = KNeighborsClassifier(n_neighbors=4)
knn1 = KNeighborsClassifier(n_neighbors=1)

knn4.fit(X_train, y_train)
knn1.fit(X_train, y_train)

y_pred4 = knn4.predict(X_test)
y_pred1 = knn1.predict(X_test)
```

```python
print("Accuracy with k=4:", accuracy_score(y_test, y_pred4)*100)
print("Accuracy with k=1:", accuracy_score(y_test, y_pred1)*100)
```

```python
plt.figure(figsize=(8, 6))
plt.subplot(1, 2, 1)
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_pred4, marker="*", s=50, edgecolors="brown")
plt.title("Predictions with k=4", fontsize=15)

plt.subplot(1, 2, 2)
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_pred1, marker="*", s=50, edgecolors="brown")
plt.title("Predictions with k=1", fontsize=15)

plt.show()
```
