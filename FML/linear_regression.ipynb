```python
# Generate data

def fake_line_with_noise(X: list[int], slope: float, intercept: float):
    y = [slope * X[i] + intercept + np.random.normal(0, 3) for i in range(len(X))]
    return y
```

```python
X = list(range(1, 11))
y = np.array(fake_line_with_noise(X, 2, 3))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("homeprices.csv")
X = df["area"]
y = df["price"]
plt.scatter(X, y)
```

```python
# Train model manually

n = len(X)
slope = ( (n * np.sum(X * y) ) - (np.sum(X) * np.sum(y))) / (
    (n * np.sum(X**2)) - (np.sum(X) ** 2)
)
intercept = ((np.sum(y) * np.sum(X**2)) - (np.sum(X) * np.sum(X * y))) / (
    (n * np.sum(X**2)) - (np.sum(X) ** 2)
)

print(slope, intercept)
plt.scatter(X, y)
plt.plot(X, slope * X + intercept, color="green")
```

```python
reshape_X = np.array(X).reshape(-1, 1)
model = LinearRegression()
model.fit(reshape_X, y)
print( model.coef_, model.intercept_ )
plt.scatter(reshape_X, y)
plt.plot(reshape_X, model.predict(reshape_X), color="green")
```
